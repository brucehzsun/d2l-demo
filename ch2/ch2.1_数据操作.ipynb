## 张量连结（concatenate）

```python
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
torch.cat((X, Y), dim=0)  # 按行拼接
```

```python
torch.cat((X, Y), dim=1)  # 按列拼接
```

```python
X == Y
```

```python
X.sum()
```

## 2.1.5. 节省内存

```python
before = id(Y)
Y = Y + X
id(Y) == before
```

```python
id(X)
```

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
before = id(X)
X += Y
id(X) == before
```
